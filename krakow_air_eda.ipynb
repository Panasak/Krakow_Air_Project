```python
# import librabies
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
# setting size for sns graph
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
# read the cleaned dataframe
df = pd.read_csv('data_sets/for_eda.csv')
```

```python
# Do pairplot (to see the relationship between each col), 
# histogram (to see distribution over time), 
# heatmap(also for corr()),
#geographic map (to see the location and the pollution level) 
df
```

```python
# pairplot (strong correlation beteewn pm1, pm2.5, pm10. More pollution in cold weather, more pollution in high air pressure (lower places))
sns.pairplot(df)
```

```python
sns.distplot(df['pm1'])
plt.savefig('data_sets/pm1_distplot.png')
```

```python
# make UTC time a datetime
df['UTC time'] = pd.to_datetime(df['UTC time'])
```

```python
# get month from datetime
df['month'] = pd.DatetimeIndex(df['UTC time']).month
```

```python
sns.barplot(x='month',y='pm1',data=df)
plt.savefig('data_sets/pm1_bar_month.png')
```

```python
sns.distplot(df['pm25'])
```

```python
sns.barplot(x='month',y='pm25',data=df)
```

```python
sns.distplot(df['pm10'])
```

```python
sns.barplot(x='month',y='pm10',data=df)
```

```python
#heatmap (There is a strong corr on the 3 particles and negative corr() to the temperature and weak possitive with the air pressure)
sns.clustermap(df.corr(), annot= True ,cmap='magma_r')
plt.savefig('data_sets/heatmap.png')
```
